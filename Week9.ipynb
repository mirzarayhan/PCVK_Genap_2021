```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

```

```python
img = cv2.imread('/content/drive/MyDrive/Images/j.png')
plt.imshow(img)
```

```python
# Structuring Element 7 x 7 tanpa dan dengan menggunakan library morphology dari openCV
# dengan 
element = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
dilate_dst = cv2.dilate(img, element)
closing_dst = cv2.erode(dilate_dst, element)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(closing_dst), plt.title('Citra Hasil Closing')
```

```python
# implementasi operasi dilasi dengan Structuring Element berbentuk Cross 3 x 3

kernel = cv.getStructuringElement(cv.MORPH_CROSS, (3, 3))
cross3 = cv.morphologyEx(img, cv.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(cross3), plt.title('Citra Hasil Dilasi 3x3')
```

```python
# implementasi operasi dilasi dengan Structuring Element berbentuk Circular 5 x 5
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5, 5))
cross3 = cv.morphologyEx(img, cv.MORPH_DILATE, kernel)
plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(cross3), plt.title('Citra Hasil Circular 5x5')
```

```python
#implementasi operasi erosi dengan Structuring Element berbentuk Rectangle 5 x 7
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 7))
cross3 = cv2.morphologyEx(img, cv2.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(cross3), plt.title('Citra Hasil Rectangle 5x7')
```
